```r
# Subset on the `$capillary_arterioles` value in the object meta data
# cells_1 = subset(pooled_ad, subset = pooled_ad$capillary_arterioles == "fetal_capillary_arterioles")
cells_1 = colnames(pooled_ad[, pooled_ad$capillary_arterioles == "fetal_capillary_arterioles"])
# cells_2 = subset(pooled_ad, subset = pooled_ad$capillary_arterioles == "organoid_capillary_arterioles")
cells_2 = colnames(pooled_ad[, pooled_ad$capillary_arterioles == "organoid_capillary_arterioles"])
```

```r
cells_1
```

```r
diff_exp_genes = MAST::MASTDETest(pooled_ad, cells_1, cells_2)
```

```r
?FindMarkers
```

```r
# Perform differential expression analysis
de_results <- FindMarkers(pooled_ad, cells_1, cells_2, test.use = "MAST", logfc.threshold = 0.25)
```

```r
de_results[order(abs(de_results[, "avg_log2FC"]), decreasing = TRUE), ]
```

```r
# Extract the differentially expressed genes
sig_genes <- rownames(de_results[de_results$p_val < 0.05, ])
```

```r
sig_genes
```

```r
# Convert the Seurat object to a matrix for input to MAST
mat <- pooled_ad@assays[rownames(pooled_ad@assays) %in% c(cells_1, cells_2), ]
```

```r
# Specify the model for MAST
mod <- model.matrix(~0 + condition, data = colData(pooled_ad@assays))
colnames(mod) <- levels(colData(pooled_ad)$condition)
```

```r
# Run MAST
fit <- mast(mat, mod)

# Results interpretation: Extract the differentially expressed genes
res <- tidy(fit)
res_sig <- res[res$qval < 0.05, ]

# Print the top 10 differentially expressed genes
head(res_sig[order(res_sig$logFC), ], 10)
```
